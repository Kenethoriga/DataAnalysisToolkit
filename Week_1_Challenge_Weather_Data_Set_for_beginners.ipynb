```python
import pandas as pd
import numpy as np
```

```python
# prompt: Find all records where the weather was exactly clear.

data[data['Weather'] == 'Clear']
```

```python
# prompt: Find the number of times the wind speed was exactly 4 km/hr.

data[data['Wind Speed_km/h'] == 4].shape[0]
```

```python
data.isnull().sum()
```

```python
# prompt: Rename the column "Weather" to "Weather_Condition."

data.rename(columns = {'Weather' : 'Weather_Condition'}, inplace = True)
data.head(2)
```

```python
# prompt: What is the mean visibility of the dataset?

data.Visibility_km.mean()
```

```python
# prompt: Find the number of records where the wind speed is greater than 24 km/hr and visibility is equal to 25 km

data[(data['Wind Speed_km/h'] > 24) & (data['Visibility_km'] == 25)].shape[0]
```

```python
numeric_cols = data.select_dtypes(include=[np.number]).columns
data.groupby('Weather_Condition')[numeric_cols].mean()
```

```python
# prompt: Find all instances where the weather is clear and the relative humidity is greater than 50, or visibility is above 40.

data[(data['Weather_Condition'] == 'Clear') & ((data['Rel Hum_%'] > 50) | (data['Visibility_km'] > 40))]
```

```python
# prompt: rename Wind Speed_km/h to Wind_speed_Kmh

data.rename(columns = {'Wind Speed_km/h' : 'Wind_speed_Kmh'}, inplace = True)
data.head(2)
```

```python
# prompt: Find the number of weather conditions that include snow

data[data['Weather_Condition'].str.contains('Snow')].shape[0]
```

```python
# prompt: Part 2: Move this CSV into a database of your choice and use SQL to answer 4 of the questions above.

import sqlite3

# Create a connection to the database (or create it if it doesn't exist)
conn = sqlite3.connect('weather_data.db')
cursor = conn.cursor()

# Create the table
cursor.execute('''
    CREATE TABLE IF NOT EXISTS weather (
        Date_Time TEXT,
        Temp_C REAL,
        Dew_Point_Temp_C REAL,
        Rel_Hum REAL,
        Wind_Speed_kmh REAL,
        Visibility_km REAL,
        Press_kPa REAL,
        Weather_Condition TEXT
    )
''')

# Insert data into the table
data.to_sql('weather', conn, if_exists='replace', index=False)

# Query 1: How many rows have 'Clear' weather?
cursor.execute("SELECT COUNT(*) FROM weather WHERE Weather_Condition = 'Clear'")
clear_count = cursor.fetchone()[0]
print("Number of 'Clear' weather rows:", clear_count)

# Query 2: What is the average Visibility_km?
cursor.execute("SELECT AVG(Visibility_km) FROM weather")
avg_visibility = cursor.fetchone()[0]
print("Average Visibility_km:", avg_visibility)

# Query 3: How many rows have Wind Speed_kmh greater than 24 and Visibility_km equal to 25?
cursor.execute("SELECT COUNT(*) FROM weather WHERE Wind_Speed_Kmh > 24 AND Visibility_km = 25")
wind_visibility_count = cursor.fetchone()[0]
print("Number of rows with high wind and low visibility:", wind_visibility_count)

# Query 4: How many rows contain 'Snow' in the Weather_Condition?
cursor.execute("SELECT COUNT(*) FROM weather WHERE Weather_Condition LIKE '%Snow%'")
snow_count = cursor.fetchone()[0]
print("Number of rows with 'Snow' in Weather_Condition:", snow_count)

# Close the connection
conn.close()
```
